## Tratando Valores Nulos

```python
def visualizar_valores_nulos(dataframe):
    # Verificar quais colunas possuem valores nulos
    colunas_nulas = dataframe.columns[dataframe.isnull().any()]

    # Iterar pelas colunas nulas e mostrar a contagem de valores nulos
    for coluna in colunas_nulas:
        contagem_nulos = dataframe[coluna].isnull().sum()
        print(f"A coluna '{coluna}' possui {contagem_nulos} valores nulos.")
```

```python
def linhas_com_valores_nulos(dataframe, coluna):
    # Filtrar as linhas que possuem valor nulo na coluna específica
    linhas_nulas = dataframe[dataframe[coluna].isnull()]

    # Filtrar as linhas com apenas um valor nulo (remover duplicatas)
    linhas_nulas_sem_duplicatas = linhas_nulas.drop_duplicates()

    return linhas_nulas_sem_duplicatas
```

```python
visualizar_valores_nulos(df_train_processed)
```

```python
linhas_com_valores_nulos(df_train_processed, 'BsmtQual')
```

#### LotFrontage trocar NaN por media, Alley por 'none',  MasVnrType remover linhas, MasVnrArea remover linhas, BsmtQual trocar por 'none', BsmtCond trocar por 'none', BsmtExposure trocar por 'none', BsmtFinType1 por 'none', BsmtFinType2 por 'none', Eletrical remover linha, FireplaceQu trocar por 'none', GarageType trocar por 'none', GarageYrBlt trocar por 0, GarageFinish trocar por 'none',GarageQual trocar por 'none', GarageCond trocar por 'none', PoolQC trocar por 'none', Fence trocar por 'none,MiscFeature trocar por 'none' 

```python
cols_to_fill_none = ['Alley', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 
                     'BsmtFinType1', 'BsmtFinType2', 'FireplaceQu', 
                     'GarageType', 'GarageFinish', 'GarageQual', 
                     'GarageCond', 'PoolQC', 'Fence', 'MiscFeature']

for col in cols_to_fill_none:
    df_train_processed[col] = df_train_processed[col].fillna('none')
    df_test_processed[col] = df_test_processed[col].fillna('none')
    
df_train_processed['GarageYrBlt'] = df_train_processed['GarageYrBlt'].fillna(0)
df_train_processed['LotFrontage'] = df_train_processed['LotFrontage'].fillna(df_train_processed['LotFrontage'].mean())

df_test_processed['GarageYrBlt'] = df_test_processed['GarageYrBlt'].fillna(0)
df_test_processed['LotFrontage'] = df_test_processed['LotFrontage'].fillna(df_test_processed['LotFrontage'].mean())

df_train_processed = df_train_processed.dropna(subset=['MasVnrType', 'MasVnrArea', 'Electrical'])
```

```python
colunas_objeto = df_test_processed.select_dtypes(include='object').columns
df_test_processed[colunas_objeto] = df_test_processed[colunas_objeto].fillna(df_test_processed[colunas_objeto].mode().iloc[0])

colunas_int = df_test_processed.select_dtypes(include='int').columns
df_test_processed[colunas_int] = df_test_processed[colunas_int].fillna(0)

colunas_float = df_test_processed.select_dtypes(include='float').columns
df_test_processed[colunas_float] = df_test_processed[colunas_float].fillna(df_test_processed[colunas_float].mean())
```

### Tratando valores categoricos

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

def encode_object_cols(df):
    # Cria uma cópia do DataFrame para não modificar o original
    df_encoded = df.copy()

    # Cria um objeto LabelEncoder
    le = LabelEncoder()

    # Itera sobre todas as colunas do DataFrame
    for col in df_encoded.columns:
        # Se o tipo da coluna é objeto (string), faz a codificação
        if df_encoded[col].dtype == 'object':
            # Transforma as categorias em números
            df_encoded[col] = le.fit_transform(df_encoded[col])

    return df_encoded
```

```python
df_test_encoded = encode_object_cols(df_test_processed)
df_train_encoded = encode_object_cols(df_train_processed)

# preciso dar fill na nas colunas object para moda e int para valor
```
